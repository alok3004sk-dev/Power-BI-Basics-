**Power BI Basics**

Q1 -  What is Power BI and why is it used in businesses?
- Answer :-
- Step 1: Introduction
Power BI is a Business Intelligence (BI) and Data Visualization tool developed by Microsoft.

- Step 2: Purpose of Power BI
It helps organizations collect, analyze, and visualize data from different sources to make better business decisions.

- Step 3: Data Connection
Power BI can connect to various data sources such as:

Excel files

SQL databases

Cloud services

Web data

CSV files
- Step 4: Data Transformation
Using Power Query, users can:

Clean data
Remove errors
Combine data from multiple sources
Transform data into a usable format
- Step 5: Data Modeling
Power BI allows users to:-
Create relationships between tables
Build calculations using DAX (Data Analysis Expressions)
Organize data efficiently
- Step 6: Data Visualization
Users can create:
Charts
Graphs
Maps
KPIs
Dashboards
Reports
- Step 7: Sharing Insights
Reports and dashboards can be published and shared with team members and management for decision-making

Q2 :- Name and explain the three main components of Power BI
- Answer:-
 - Three Main Components of Power BI
- Power BI Desktop - Used to connect, transform, analyze data, and create reports.
- Power BI Service - Cloud-based platform used to publish, share, and collaborate on reports and dashboards.
- Power BI Mobile - Mobile app used to view and monitor reports and dashboards on smartphones and tablets.

Q3 :- Explain the Power BI workflow
- Answer
- Connect Data – Import data from sources like Excel, SQL Server, or databases.
- Transform Data – Clean and prepare the data using Power Query.
- Model Data – Create relationships and calculations between tables.
- Create Reports – Build charts, graphs, and visualizations.
- Publish Reports – Upload reports to Power BI Service.
- Share & Monitor – Share dashboards with users and track business performance.

Q4 :- List any four data cleaning tasks that can be performed in Power Query.
- Answer:-
- Four Data Cleaning Tasks in Power Query
- Remove Duplicates – Eliminates repeated records from the dataset.
- Remove Null/Blank Values – Deletes or replaces missing data.
- Change Data Types – Converts data into the correct format (Text, Number, Date, etc.).
- Rename Columns – Changes column names to make them more meaningful and understandable

Q5:-  Write step by step instructions to load the above dataset.
- Answer:-
- Step-by-Step Instructions to Load the Dataset in Power BI
- Open Power BI Desktop.
- Click Home → Get Data.
- Select Excel Workbook and click Connect.
- Browse to the location of the Sample Real_Estate Data file.
- Select the file and click Open.
In the Navigator window, select the required sheet(s) (e.g., Orders, Returns, People).
- Click Load to import the data directly into Power BI.
Or click Transform Data if you want to clean or modify the data before loading.
- Wait for the data to load.
- Verify that the tables appear in the Fields pane on the right side of the screen.
- The dataset is now ready for data modeling and report creation.

- Flow:
Open Power BI → Get Data → Excel Workbook → Select File → Choose Sheets → Load Data → Start Analysis.

Q8 :- Split Owner Name to create two new columns as First Name and Last Name
- Answer:-
- Split Owner Name into First Name and Last Name in Power BI

Step 1: Open Power Query Editor.

Step 2: Select the Owner Name column.

Step 3: Go to Transform → Split Column → By Delimiter.

Step 4: Choose Space as the delimiter.

Step 5: Select Split at the left-most delimiter and click OK.

Step 6: Rename the new columns:

Column 1 → First Name
Column 2 → Last Name

Result: The Owner Name column is split into First Name and Last Name columns.
